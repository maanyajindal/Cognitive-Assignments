```python
def swap_list_elements(lst, i, j):
    temp = lst[i]
    lst[i] = lst[j]
    lst[j] = temp
    return lst

my_list = [1, 2, 3, 4, 5]
print("Swapped List:", swap_list_elements(my_list, 1, 3))
```

```python
def swap_elements_in_set(s, elem1, elem2):
   # """Swaps two elements in a set by converting it to a list first."""
    if elem1 not in s or elem2 not in s:
        print("Both elements must be in the set.")
        return s
    
    lst = list(s)
    
    idx1, idx2 = lst.index(elem1), lst.index(elem2)

    lst[idx1], lst[idx2] = lst[idx2], lst[idx1]

    return set(lst)

my_set = {10, 20, 30, 40, 50}
print("Original set:", my_set)

swapped_set = swap_elements_in_set(my_set, 20, 50)
print("Set after swapping:", swapped_set)
```
