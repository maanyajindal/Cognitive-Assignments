```python
#1
import numpy as np
import matplotlib.pyplot as plt
M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10,20)
y1 = M * x ** 2
y2 = M * np.sin(x)
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='y = M * x^2', linestyle='--', color='green')
plt.plot(x, y2, label='y = M * sin(x)', linestyle='-', color='red')
plt.legend()
plt.grid()
plt.title('Maths Fx Plots')
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
#2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.DataFrame({'Subjects': ['Math', 'Sci', 'Eng', 'sst', 'hin'],
                   'Scores': [85, 78, 90, 88, 95]})
plt.figure(figsize=(8, 5))
sns.barplot(x='Subjects', y='Scores', data=df, palette='viridis')
for i, score in enumerate(df['Scores']):
    plt.text(i, score + 1, str(score), ha='center', fontsize=12)
plt.title('Subject Scores')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(axis='y')
plt.show()
```

```python
#3
import numpy as np
import matplotlib.pyplot as plt
roll_number = int(input("Enter your roll number as a seed: "))
np.random.seed(roll_number)
data = np.random.randn(50)
print(data)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(np.cumsum(data), color='blue')
axes[0, 0].set_title("Cumulative Sum")
axes[0, 0].grid()

axes[0, 1].scatter(range(50), data, color='red')
axes[0, 1].set_title("Scatter Plot with Noise")
axes[0, 1].grid()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url = 'https://github.com/AnjulaMehto/MCA/raw/main/company_sales_data.csv'
data = pd.read_csv(url)
data['month_number'] = data['month_number'].astype(int)
plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=data, marker='o', color='b')
plt.title('Total Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.xticks(data['month_number'])  
plt.show()
plt.figure(figsize=(12, 8))
product_columns = data.columns[1:7]  # Assuming the first column is 'month_number' and the next six are product sales
for product in product_columns:
    sns.lineplot(x='month_number', y=product, data=data, marker='o', label=product)
plt.title('Monthly Sales Data for All Products')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.grid(True)
plt.xticks(data['month_number'])
plt.legend(title='Products')
plt.show()

# 3. Bar Chart: All Features/Attributes
# Melt the DataFrame to have 'month_number' and 'variable' columns for plotting
melted_data = pd.melt(data, id_vars=['month_number'], var_name='Attribute', value_name='Value')
plt.figure(figsize=(14, 10))
sns.barplot(x='month_number', y='Value', hue='Attribute', data=melted_data)
plt.title('Bar Chart of All Features/Attributes per Month')
plt.xlabel('Month Number')
plt.ylabel('Value')
plt.grid(True)
plt.xticks(data['month_number'])
plt.legend(title='Attributes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
